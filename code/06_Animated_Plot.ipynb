```python
## Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from PIL import Image, ImageOps
from IPython.display import HTML
```

```python
TRACK_HEIGHT = 820
TRACK_WIDTH = 965
fastest_lap = 110

# Paths
csv_file = '../data/11_12_2021/19_28_18/fast_laps.csv'
img_file = '../assets/images/grass_track.png'
```

```python
df = pd.read_csv(csv_file)
```

This loads the data from the fastest lap (calculated previously) and plots it. It takes a couple of minutes to run.

```python
# https://stackoverflow.com/questions/25333732/matplotlib-animation-not-working-in-ipython-notebook-blank-plot
df['shift_x'] = df['pos_x'] - df['pos_x'].min()
df['shift_z'] = df['pos_z'] - df['pos_z'].min()
df['scaled_x'] = df['shift_x'] * (TRACK_WIDTH / df['shift_x'].max())
df['scaled_z'] = df['shift_z'] * (TRACK_HEIGHT / df['shift_z'].max())
# Hand tuned, will fix later.
x = (df.loc[df['lap']==110,'scaled_x'] * 0.975) + 15
y = (df.loc[df['lap']==110,'scaled_z'] * 0.975) + 20

img = Image.open(img_file).copy()
img = img.transpose(method=Image.FLIP_TOP_BOTTOM)

plt.rcParams["animation.html"] = "html5"
fig, ax = plt.subplots(figsize=(12,12), constrained_layout=True)
line, = ax.plot(x, y, color='deeppink', linewidth=3, linestyle=(0, (2,1)))
ax.axis('off')
ax.set_aspect('equal')

plt.imshow(img);

ax.set_ylim(img.height, 0)

def update(num, x, y, line):
    line.set_data(x[:num], y[:num])
    line.axes.axis([0, img.width, 0, img.height])
    return line,

ani = animation.FuncAnimation(fig, update, len(x), fargs=[x, y, line],
                              interval=16, blit=True, repeat=True)
```
